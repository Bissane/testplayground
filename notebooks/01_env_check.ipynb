```python
import chromadb, os
os.makedirs("data/chroma", exist_ok=True)
client = chromadb.PersistentClient(path="data/chroma")  # keeps data on disk
col = client.get_or_create_collection("toy_docs")
```

```python
from sentence_transformers import SentenceTransformer
m = SentenceTransformer("all-MiniLM-L6-v2")

docs = [
    "Transformers are great for NLP.",
    "Apples grow on trees.",
    "Stockholm is lovely in summer."
]
ids = [f"doc-{i}" for i in range(len(docs))]
embs = m.encode(docs, convert_to_numpy=True).tolist()

col.upsert(documents=docs, embeddings=embs, ids=ids)

q = "Which sentence mentions Sweden?"
res = col.query(query_texts=[q], n_results=2)
res
```
